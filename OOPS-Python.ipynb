### decorators

```python
def main_welcome(msg):

    def sub_main_class():
        print("loda mera")
        print(msg)

    return sub_main_class()


##this is an examlple of closure where anything inside the main function will be available to the child function
```

```python
main_welcome("mcmc")
```

```python
def main_welcome(func):

    def sub_main_class():
        print("loda mera")
        func("kya kar rha hai bsdk")

    return sub_main_class()


##this is an examlple of closure where anything inside the main function will be available to the child function
```

```python
main_welcome(print)
```

```python
def main_welcome(func):

    def sub_main_class():
        print("loda mera")
        func()

    return sub_main_class()
```

```python
@main_welcome
def func():
    print("hogaya")   ##decorator
```

```python
main_welcome(func)
```

## static method

## why to use
very fast
gets initialized as soon as the class is loaded
gets initilized only once in the entire func call

```python
import datetime
now = datetime.datetime.now()
```

```python
class Car:
    base_price = 10000
    def __init__(self,windows,doors,power):
        self.windows = windows
        self.doors = doors
        self.power = power
    def __str__(self):
        return "the base price is {}".format(self.base_price)
    @classmethod
    def revise_base_price(cls,inflation):
        cls.base_price += inflation*cls.base_price
    @staticmethod
    def check_year():
        if now.year ==  2024:
            return True
        else:
            return False
```

## multiple inheritance

```python
class A:
    def method1(self):
        print("method 1 is called")
```

```python
class B(A):
    def method1(self):
        print("B method 1 is called")
```

```python
class C(A):
    def method1(self):
        print("C method 1 is called")
```

```python
class D(B,C):
    def method1(self):
        print("D method1 is called")
```

```python
d = D()
```

```python
d.method1()
```

```python
B.method1(d)
```

```python
A.method1(d)
```

```python
class D(B,C):
    def method1(self):
        print("D method1 is called")
        C.method1(self)
        B.method1(self)
        D.method1(self)
```

```python
d = D()
```

```python
d.method1()
```
